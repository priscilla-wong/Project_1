```python
# Imports

import panel as pn
pn.extension('plotly')
import plotly.express as px
import pandas as pd
import hvplot.pandas
import matplotlib.pyplot as plt
import os
from pathlib import Path
from dotenv import load_dotenv
import numpy as np
%matplotlib inline
import seaborn as sns
```

```python
# Read in data and format for PCE

# Bring in covid data and group for total cases per month

covid_path = Path("../Resources/covid_data.csv")
covid_data = pd.read_csv(covid_path)
covid_df = covid_data.drop(columns=["year", "month", "day", "conf_cases", "prob_cases", "pnew_case", "conf_death", "prob_death", "new_death", "pnew_death", "created_at", "consent_cases", "consent_deaths"])
covid_df = covid_df.sort_values(by=['state', 'yearmonth'])
covid_df = covid_df.groupby('yearmonth')['new_case'].sum()

# Bring in covid data and group for average new cases per month

covid_path = Path("../Resources/covid_data.csv")
covid_data = pd.read_csv(covid_path)
covid_df_avg = covid_data.drop(columns=["year", "month", "day", "conf_cases", "prob_cases", "pnew_case", "conf_death", "prob_death", "new_death", "pnew_death", "created_at", "consent_cases", "consent_deaths"])
covid_df_avg = covid_df_avg.groupby('yearmonth')['new_case'].mean()
columns=['yearmonth', 'Average New Cases']
covid_df_avg.columns = columns

# Bring in PCE data

pc_path = Path("../Resources/pc_data.csv")
pc_data = pd.read_csv(pc_path)
pc_df = pc_data.dropna()
pc_df.rename(columns={pc_df.columns[1]: "Category"}, inplace = True)
pc_df = pc_df.drop(columns=pc_df.columns[0]).copy()
pc_df = pc_df.drop(columns=pc_df.columns[1]).copy()
pc_df.reset_index()
pc_df = pc_df.drop(pc_df.columns.to_series()['1959M01':'2019M12'], axis=1)
pc_df = pc_df.transpose()
pc_df = pc_df.iloc[:,0]

# Concatenate data

cov_pc_df = pd.concat([covid_df, covid_df_avg, pc_df], axis="columns", join="inner")
cov_pc_df = cov_pc_df.reset_index()
columns = ['Period','Total Covid Cases', 'Average New Cases', 'Personal Consumption Expenditure']
cov_pc_df.columns = columns
cov_pc_df['Personal Consumption Expenditure'] = cov_pc_df['Personal Consumption Expenditure'].str.replace(',','')
cov_pc_df['Personal Consumption Expenditure'] = cov_pc_df['Personal Consumption Expenditure'].astype(int)

# Correlation for data

cov_pc_corr = cov_pc_df.corr()
```

```python
# Define objects for PCE plots

cov_pc_scatter = cov_pc_df.plot(kind='scatter',
                                 x='Total Covid Cases',
                                 y='Personal Consumption Expenditure',
                                 title='Relationship Between Covid Cases and Personal Consumption Expenditure'
                                )

pc_line = cov_pc_df.plot(title='Personal Consumption from Jan 2020 Through Feb 2021',
                         rot=90,
                         y='Personal Consumption Expenditure',
                         x='Period'
                        )

cov_pc_bar = cov_pc_df.hvplot.bar('Period',
                                  ['Total Covid Cases', 'Personal Consumption Expenditure'],
                                  title='Covid Case Effects on Personal Consumption',
                                  xlabel='Period',
                                  ylabel='Covid Cases and Expenditure',
                                  rot=90,
                                  width=900,
                                  height=500
                                 )

cov_pc_heat = sns.heatmap(cov_pc_corr, vmin=-1, vmax=1)
```
